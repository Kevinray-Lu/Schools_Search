```python
# get all information of Chinese middle/high schools
import requests
from bs4 import BeautifulSoup
import csv
def one_page(page_number):
    # get the website that has all information needed
    html = "http://xuexiao.51sxue.com/slist/?t=3&areaCodeS=&page=" + str(page_number)
    res = requests.get(html)
    # change encoding to read Chinese characters
    res.encoding ="gb18030"
    soup = BeautifulSoup(res.text, "html.parser")
    #get the tag that contains school names
    every = soup.find_all("h3")
    school_names = []
    count = 0
    #fetch school names
    for items in every:
            if count == 10:
                break;
            school_names.append(items.text)
            count += 1
    # get tags that contains all school information
    every2 = soup.find_all("li")
    count2 = 0
    complete_list = []
    temp = []
    for items in every2:
        if len(school_names) == 0:
            break
        # find school information according to saved school names, replace special characters with space
        if school_names[count2] in items.text  :
            temp = [school_names[count2].replace("\xa0"," ")]
        elif count2 < len(school_names)-1 and school_names[count2+1] in items.text :
            count2 += 1
            temp = [school_names[count2].replace("\xa0"," ")]
        # fetch location, address, property, contact information of schools
        elif "地区" in items.text or "学校地址" in items.text or "属性" in items.text or "性质" in items.text:
            temp.append(items.text.replace('\n'," ").replace("\u2003", " ").replace("\ue003"," ").replace("\u200e"," ").replace('\xa0'," "))
        elif "联系电话" in items.text:
            temp.append(items.text.replace('\xa0'," ").replace("\u2003", " ").replace("\ue003"," "))
            complete_list.append(temp)
            if count2 == len(school_names)-1:
                break
        else:
            continue
    # save all school information to csv
    for items in complete_list:
        print(items)
        with open('工作簿3.csv',mode='a',newline = "") as csv1:
            #if items[0] == "北京市第三十中学":
                #continue        
            csv_writer = csv.writer(csv1, delimiter = ",")
            csv_writer.writerow(items)
     
```

```python
# show some examples
i = 1077
while i <= 1338:
    one_page(i)
    i+=1
    print(i)
```

```python
# get all information of US colleges
import requests
from bs4 import BeautifulSoup
import csv
def US_UNI(page):
    # get the website with all information needed
    html = 'https://api.data.gov/ed/collegescorecard/v1/schools.json?fields=school.name,school.state,school.city,school.alias,&per_page=1000&api_key=UP2zqmzJ7CwWToqgbfGJ1XNku6VL9cznUYhQCS0d&page='+ str(page) 
    res = requests.get(html)
    res.encoding ="gb18030"
    # get rid of brackets
    res_final = res.text.split("[")[1]
    res_final = res_final.split("}")
    every_page = []
    for items in res_final:
        # skip head
        if res_final.index(items) > 0:
            items = items[1:]
        # get rid of excess punctuations
        dealt = items.replace('"', "").replace("{", "").split(",")
        # initialize temp list to store information about one college
        final_list = ["","","",""]
        # get all information needed
        for i in dealt:
            if "name" in i:
                final_list[0] = i
            elif "state" in i:
                final_list[1] = i
            elif "city" in i:
                final_list[2] = i
            elif "alias" in i:
                final_list[3] = i
        every_page.append(final_list)
    # save information to csv file
    for items in every_page:
        #with open('工作簿1.csv',mode='a',newline = "") as csv1:   
            #csv_writer = csv.writer(csv1, delimiter = ",")
            #csv_writer.writerow(items)
            print(items)
```
